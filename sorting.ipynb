### Bubble Sort
#### Time = O(N^2)
#### Space = O(1)

```python
#Bubble Sort

A = [-5, 3, 2, 1, -3, -3, 7, 2, 2]

def bubbleSort(arr):
    n= len(arr)
    flag = True
    while flag:
        flag = False 
        for i in range(1,n):
            if arr[i-1] > arr[i]:
                #set flag to True and swap
                flag = True
                arr[i], arr[i-1] = arr[i-1], arr[i]
                
bubbleSort(A)
print(A)
```

### Insertion Sort
#### Time = O(N^2)
#### Space = O(1)

```python
#Insertion Sort
def insertionSort(arr):
    n = len(arr)
    for i in range(1,n):
        for j in range(i,0,-1):
            #if not in order, do swap
            if arr[j-1]>arr[j]:
                arr[j-1], arr[j] = arr[j], arr[j-1]
            #else, section is in order break
            else:
                break
            
insertionSort(A)
print(A)
```

### Selection Sort
#### Time = O(N^2)
#### Space = O(1)

```python
def selectionSort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range (i+1,n):
            if arr[j]<arr[i]:
                min_idx = j
        #swap arr[i] with arr[min_idx]
        arr[i],arr[min_idx] = arr[min_idx],arr[i]
        
selectionSort(A)
print(A)
```

### Merge Sort
#### Time = O(NLogN)
#### Space = O(N)

```python
#merge sort

def mergeSort(arr):
    n= len(arr)
    
    if n==1:
        return arr
    
    m=len(arr)//2
    #split based on m
    L = arr[:m]
    R = arr[m:]
    
    L = mergeSort(L)
    R = mergeSort(R)
    l,r=0,0
    
    L_len = len(L)
    R_len = len(R)
    
    #merge step
    sorted_arr = [0]*n
    i=0
    
    #leniar time merge of sorted array
    while l<L_len and r<R_len:
        if L[l] < R[r]:
            sorted_arr[i] = L[l]
            l+=1
        else:
            sorted_arr[i] = R[r]
            r+=1
        i+=1
        
    #Either one of the above cases must have gone out of bound to break the loop so we need to fill in for the the index that didnt go out of bounds
    while l<L_len:
        sorted_arr[i] = L[l]
        l+=1
        i+=1
        
    while r<R_len:
        sorted_arr[i] = R[r]
        r+=1
        i+=1
        
    return sorted_arr
        
mergeSort(A)
print(A)
```

```python
#QuickSort

def quickSort(arr):
    if len(arr)<=1:
        return arr
    
    p = arr[-1]
    L = [x for x in arr[:-1] if x<=p]
    R = [x for x in arr[:-1] if x>p]
    
    L = quickSort(L)
    R = quickSort(R)
    
    return L + [p] + [R]

quickSort(A)
print(A)
```

### Counting Sort
#### Time = O(N+K)
#### Space = O(K)

```python
#counting sort
F = [8,5,9,1,3,4,5,10,15,22,8,7,11,2]

def countingSort(arr):
    n = len(arr)
    maxx = max(arr)
    counts = [0]*(maxx+1)
    
    for x in arr:
        counts[x]+=1
        
    i = 0
    for c in range(maxx+1):
        while counts[c]>0:
            arr[i]=c
            i+=1
            counts[c]-=1
            
countingSort(F)
print(F)
    
```
