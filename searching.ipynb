### Naive Search
#### Time = O(N)
#### Space = O(1)

```python
A = [-3, -1, 0, 1, 4, 7, 8]

if 8 in A:
    print(True)
```

### Binary Search
#### Time = O(logN)
#### Space = O(N)

```python
def binarySearch(arr, target):
    n = len(arr)
    l=0
    r=n-1
    while l<=r:
        m = l+((r-l)//2)
        
        if arr[m] == target:
            return True
        elif target<arr[m]:
            r = m-1
        else:
            l = m+1

    return False

binarySearch(A,18)
```

### More efficient Search
#### Time = O(logN)
#### Space = O(N)

```python
B = [False, False, False, True, True, True, True]

def binarySearchCondition(arr):
    
    n = len(arr)
    l=0
    r=n-1
    
    while l<r:
        m = (l+r)//2
        if B[m]:
            r=m
        else:
            l=m+1
        
    return l #or r either works
    
binarySearchCondition(B)
    
```
